# 🔰 1. What is SQL?
SQL (Structured Query Language) is the standard language for interacting with relational databases. It lets you store, retrieve, update, and delete data. Popular databases: MySQL, PostgreSQL, SQLite, SQL Server, Oracle.

**SQL (Structured Query Language) is used to:**
- Store data
- Retrieve data
- Manage databases

👉 Used in:

- Web apps (backend)
- Data analysis
- Data science
- Backend APIs
